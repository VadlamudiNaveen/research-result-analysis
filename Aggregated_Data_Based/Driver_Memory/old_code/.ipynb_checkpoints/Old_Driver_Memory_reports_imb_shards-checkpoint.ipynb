```python
# Optimized parameters from chatgpt
import numpy as np
import matplotlib.pyplot as plt

# Sample data for result_datastructure
result_datastructure = {
    '4GB': (312.2, 345.7, 0.0, 0.0),
    '8GB': (72.8, 78.9, 63.9, 66.7),
    '16GB': (72.5, 70.1, 61.1, 64.9),
    '32GB': (68.3, 76.6, 64.9, 68.8)
}

batches = ("n = 2", "n = 3", "n = 4", "n = 5")
x = np.arange(len(batches))  # the label locations

# Set the width of each bar to 0.6
width = 0.6

fig, ax = plt.subplots()

colors = ['royalblue', 'forestgreen', 'orangered', 'goldenrod']
hatch_patterns = ['', '/', '\\', '.']

# Calculate the total width for all bars and the spacing between each group
total_width = width * len(result_datastructure)
spacing = 0.2

# Calculate the x positions for each group of bars
x_positions = np.arange(len(batches)) * (total_width + spacing )

for idx, (attribute, values) in enumerate(result_datastructure.items()):
    rects = ax.bar(x_positions + idx * width, values, width, label=attribute, color=colors[idx], hatch=hatch_patterns[idx])
    ax.bar_label(rects, padding=3)  # adjust padding for bar labels

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Time taken to Execute in seconds')
ax.set_title('Effect of Driver Memory on Algorithms')
ax.set_xticks(x_positions + (total_width - width) / 2)
ax.set_xticklabels(batches)
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))  # adjust legend position
ax.set_ylim(0, 500)  # adjust y-axis limit based on data

plt.show()
```
